# BookMyShow (Latest Movies in Theater Near Me)

```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By

service = Service('chromedriver.exe')
options = Options()

driver = webdriver.Chrome(service=service, options=options)
driver.get('https://in.bookmyshow.com/explore/movies-chandigarh')

movie = driver.find_element(by=By.CSS_SELECTOR, value='.sc-b1h692-8.kgjaGy')

movie_grid = movie.find_elements(by=By.CSS_SELECTOR, value='.sc-1ljcxl3-0.ldQqlW')

Movie = []
Links = []
Age_Rating = []
Languages = []
Meta_data = []

for movie in movie_grid:
    
    for title in movie.find_elements(By.CSS_SELECTOR,'.sc-7o7nez-0.elfplV'):
        Movie.append(title.text)

    for meta_data in movie.find_elements(By.CSS_SELECTOR, '.sc-133848s-2.sc-133848s-12.ccqrhI.jgYpvq'):
        for data in meta_data.find_elements(By.CSS_SELECTOR, '.sc-7o7nez-0.bsZIkT'):
            Meta_data.append(data.text)

    for poster in movie.find_elements(By.CSS_SELECTOR, '.sc-133848s-11.sc-1ljcxl3-1.ctsexn.uPavs'):             
        Links.append(poster.get_attribute('href'))
        

for i in range(0,len(Meta_data), 2):
    Age_Rating.append(Meta_data[i])
    Languages.append(Meta_data[i+1])

BookMyShowMovie = pd.DataFrame({"Movie":Movie, "AgeRating":Age_Rating,  "Languages":Languages, "Link":Links})

driver.close()
```
