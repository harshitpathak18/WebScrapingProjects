# IMDB (Movies In Theaters)

```python
import time
import numpy as np
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By

service = Service('chromedriver.exe')
options = Options()

driver = webdriver.Chrome(service=service, options=options)
driver.get('https://www.imdb.com/showtimes/?ref_=nv_mv_sh')

movie_list = driver.find_element(by=By.CSS_SELECTOR, value='ul[data-testid="movies-near-you-list"]')


data = []
for movie in movie_list.find_elements(By.CLASS_NAME,'ipc-metadata-list-summary-item')[0:25]:
    movie_name = movie.find_element(By.CLASS_NAME, 'ipc-title__text').text
    poster = movie.find_element(By.CLASS_NAME, 'ipc-image').get_attribute('src')
    link = movie.find_element(By.CSS_SELECTOR, '.ipc-lockup-overlay.ipc-focusable').get_attribute('href')

    meta_data = movie.find_element(By.CSS_SELECTOR, 'div[data-testid="ratingGroup--container"]').text
    meta_data = meta_data.split('\n')

    movie_info = movie.find_element(By.CSS_SELECTOR, '.sc-4b408797-7.fUdAcX.mny-title-metadata').text
    movie_info = movie_info.split('\n')[0:3]
    
    data.append({
        "Movie": movie_name,
        "Rating": meta_data[0],
        "TotalRating": meta_data[1],
        "Year": movie_info[0],
        "Duration":movie_info[1],
        "Poster":poster,
        "Link":link,
    })


imdb_movie = pd.DataFrame(data)

driver.close()
```
