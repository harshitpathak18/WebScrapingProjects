# Rotten Tomato 

## Movies

```python
import time
import numpy as np
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By

service = Service('chromedriver.exe')
options = Options()

driver = webdriver.Chrome(service=service, options=options)
driver.get('https://www.rottentomatoes.com/browse/movies_in_theaters/sort:popular')


movie_list = driver.find_elements(by=By.CLASS_NAME, value='flex-container')

rows = []
for movie in movie_list:
    movie_name=movie.find_element(By.CSS_SELECTOR, 'span[data-qa="discovery-media-list-item-title"]').text
    poster = movie.find_element(By.CLASS_NAME, 'posterImage').get_attribute('src')
    release_date = movie.find_element(By.CSS_SELECTOR,'span[data-qa="discovery-media-list-item-start-date"]').text
        
    rows.append({
        "Movie":movie_name,
        "Release":release_date,
        "Poster":poster
    })
 
rotten_movie = pd.DataFrame(rows)

driver.close()
```

## Shows

```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By


service = Service('chromedriver.exe')
options = Options()

driver = webdriver.Chrome(service=service, options=options)
driver.get('https://www.rottentomatoes.com/browse/tv_series_browse/')


movie = driver.find_element(by=By.CLASS_NAME, value='discovery-tiles__wrap')

Poster = []
for poster in movie.find_elements(by=By.CSS_SELECTOR, value='rt-img[class="posterImage"]'):
    Poster.append(poster.get_attribute('src'))

Show = []
for title in movie.find_elements(by=By.CSS_SELECTOR, value='span[data-qa="discovery-media-list-item-title"]'):
    Show.append(title.text)

CriticScore = []
for score in movie.find_elements(by=By.CSS_SELECTOR, value='rt-text[slot="criticsScore"]'):
    CriticScore.append(score.text)

ReleaseDate = []
for release in movie.find_elements(by=By.CSS_SELECTOR, value='span[data-qa="discovery-media-list-item-start-date"]'):
    ReleaseDate.append(release.text.replace('Opened ',''))
 
rotten_show = pd.DataFrame({"Show":Show, "CriticScore":CriticScore, "Poster":Poster})

driver.close()
```
