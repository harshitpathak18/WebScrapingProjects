```python
import time
import numpy as np
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

service = Service('chromedriver.exe')
options = Options()

driver = webdriver.Chrome(service=service, options=options)
url = 'https://www.booking.com/searchresults.html?ss=New%20Delhi%2C%20Delhi%20NCR%2C%20India&ssne=Chand%C4%ABgarh&ssne_untouched=Chand%C4%ABgarh&aid=355028&lang=en-us&sb=1&src_elem=sb&src=searchresults&dest_id=-2106102&dest_type=city&place_id=city%2F-2106102&ac_position=0&ac_click_type=b&ac_langcode=en&ac_suggestion_list_length=5&search_selected=true&search_pageview_id=4e2b2816278d011e&ac_meta=GhA0ZTJiMjgxNjI3OGQwMTFlIAAoATICZW46A0RlbEAASgBQAA%3D%3D&checkin=2025-05-08&checkout=2025-05-09&group_adults=2&no_rooms=1&group_children=0'
driver.get(url)


# -------------------------------INFINITE SCROLL----------------------------------------------
wait = WebDriverWait(driver, 10)

height = driver.execute_script("return document.body.scrollHeight")

while True:
    driver.execute_script("window.scrollTo(0,document.body.scrollHeight)")
    time.sleep(2)
    new_height = driver.execute_script("return document.body.scrollHeight")
    if height==new_height:
        break

    height=new_height

# --------------------------------------------------------------------------------

# Extract hotel cards
hotel_cards = wait.until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, 'div[data-testid="property-card"]')))

rows = []

for hotel in hotel_cards:
    hotel_name = hotel.find_element(By.CSS_SELECTOR, 'div[data-testid="title"').text
    location = hotel.find_element(By.CSS_SELECTOR, 'span[data-testid="address"]').text
    link = hotel.find_element(By.CSS_SELECTOR, 'a[data-testid="title-link"]').get_attribute('href')

    distance_from_downtown = hotel.find_element(By.CSS_SELECTOR, 'span[data-testid="distance"]').text
    
    price = int(
        hotel.find_element(By.CSS_SELECTOR, 'span[data-testid="price-and-discounted-price"]').text
        .replace(',', '').replace('₹','')
    )
    
    try:
        review_block = hotel.find_element(By.CSS_SELECTOR, 'div[data-testid="review-score"]')
        
        try:
            rating = float(review_block.find_element(By.CSS_SELECTOR, 'div[aria-hidden="true"]').text)     
        except:
            rating = np.nan        
    except:
        pass

    rows.append({'Hotel' : hotel_name, 
                 'Locality' : location, 
                 "Downtown(km)" : distance_from_downtown,
                 'Rating' : rating,
                 'Price(₹)' : price,
                 'Link' : link
                })

hotel_data = pd.DataFrame(rows)

driver.close()
```

```python
hotel_data.to_csv('NewDelhi_Hotel_List.csv', index = False)
```
