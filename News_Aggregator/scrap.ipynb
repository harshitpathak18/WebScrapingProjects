# Indian Express

```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service

service = Service('chromedriver.exe')

options = Options()
options.add_argument('--headless=new')

driver = webdriver.Chrome(service=service, options=options)
driver.get('https://indianexpress.com/latest-news/?ref=latestnews')

articles = driver.find_elements(by=By.CLASS_NAME, value='articles')

Headline = []
Summary = []
Source = []
for article in articles:
    Headline.append(article.find_element(by=By.CLASS_NAME, value='title').text)
    Summary.append(article.find_element(by=By.TAG_NAME, value='p',).text)
    Source.append('The Indian Express')

indian_express = pd.DataFrame({"Source":Source, "Headline":Headline, "Summary":Summary})
driver.close()
```

# NBC News

```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service

service = Service('chromedriver.exe')

options = Options()
options.add_argument('--headless=new')

driver = webdriver.Chrome(service=service, options=options)
driver.get('https://www.nbcnews.com/')

headlines = driver.find_elements(by=By.CLASS_NAME, value='headline-large')
summaries = driver.find_elements(by=By.CSS_SELECTOR, value='.multi-story__dek.publico-txt.f3.lh-copy.fw4')

Headline = []
Summary = []
Source = []
for headline, summary in zip(headlines, summaries):
    Headline.append(headline.text)
    Summary.append(summary.text)
    Source.append('NBC')

nbc = pd.DataFrame({"Source":Source, "Headline":Headline, "Summary":Summary})
driver.close()
```

# Times of India

```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service

service = Service('chromedriver.exe')

options = Options()
options.add_argument('--headless=new')

driver = webdriver.Chrome(service=service, options=options)
driver.get('https://timesofindia.indiatimes.com/briefs')

articles = driver.find_elements(by=By.CLASS_NAME, value='brief_box')

Headline = []
Summary = []
Source = []
for article in articles:
    Headline.append(article.find_element(by=By.TAG_NAME, value='h2').text)
    Summary.append(article.find_element(by=By.TAG_NAME, value='p').text)
    Source.append('Times of India')

times_of_india = pd.DataFrame({"Source":Source, "Headline":Headline, "Summary":Summary})
driver.close()
```

```python
times_of_india.head()
```

# Aggregating all the news

```python
aggregated_news = pd.concat([indian_express, nbc, times_of_india], ignore_index=True)

aggregated_news
```
